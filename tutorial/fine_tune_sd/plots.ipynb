```python
import matplotlib.pyplot as plt

# Sample data (replace this with your own data)
categories = [' A', 'Category B', 'Category C', 'Category D', ' ', ' B']
percentages = [100, 80, 60, 40, 20, 0]

# Calculate the total percentage for normalization
total_percentage = sum(percentages)

# Calculate the percentage of each category
normalized_percentages = [percentage / total_percentage * 100 for percentage in percentages]

# Create a horizontal bar plot
plt.barh(categories, normalized_percentages, color='skyblue')

# Add labels and title
plt.xlabel('Percentage (%)')
plt.ylabel('Categories')
plt.title('Title')

# # Display percentages on the bars
# for index, value in enumerate(normalized_percentages):
#     plt.text(value, index, f'{value:.1f}%', va='center')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = np.arange(0, 250, 50)
weight_counts = {
    "Below": np.array([100, 75, 50, 25, 0]),
    "Above": np.array([0, 25, 50, 75, 100]),
}
height = 20.5

fig, ax = plt.subplots()
left = np.zeros(5)

for boolean, weight_count in weight_counts.items():
    p = ax.barh(species, weight_count, height, label=boolean, left=left)
    left += weight_count

ax.set_title("Title")
ax.set_xlabel("Lora Scale")

ax.set_ylabel("Epochs")
ax.set_yticks(species)

plt.show()
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from glob import glob

image_files = glob("./gifs/imgs/*.png")[:5]
# Define image file paths


# Create subplots for each image vertically
fig, axs = plt.subplots(5, 1, figsize=(6, 12))

# Iterate over image files and plot them
for i, image_file in enumerate(image_files):
    # Load and display the image
    img = mpimg.imread(image_file)
    axs[i].imshow(img)
    axs[i].axis('off')  # Turn off axis ticks and labels

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg



# Create subplots for the main plot and the side plot
fig, (ax_main, ax_side) = plt.subplots(1, 2, figsize=(12, 6))

# Main plot (line plot as an example)
x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values)
ax_main.plot(x_values, y_values)
ax_main.set_title('Main Plot')

# Side plot with vertical arrangement of images
for i, image_file in enumerate(image_files):
    # Load and display the image
    img = mpimg.imread(image_file)
    ax_side.imshow(img)
    ax_side.axis('off')  # Turn off axis ticks and labels
    if i < len(image_files) - 1:
        ax_side = plt.subplot(1, 2, 2)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
